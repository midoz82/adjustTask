```python
import sqlalchemy as db
from sqlalchemy import orm
from sqlalchemy import text
import pandas as pd
```

```python
engines = {
    "db1": db.create_engine("postgres://localhost:5433/db-1"),
    "db2": db.create_engine("postgres://localhost:5434/db-2"),
    "db3": db.create_engine("postgres://localhost:5435/db-3")
} 
```

```python
sessions = {
    "session1": orm.sessionmaker(engines["db1"]),
    "session2": orm.sessionmaker(engines["db2"]),
    "session3": orm.sessionmaker(engines["db3"])
} 
```

```python
sql = text("select country, sum(installs) from installs_by_country where EXTRACT(Month from created_at) = 5 and EXTRACT(Year from created_at) = 2019 and paid = true group by country")
results = []
results.append( engines["db1"].execute(sql))
results.append( engines["db2"].execute(sql))
results.append( engines["db3"].execute(sql))
data=[]
for result in results:
    for row in result:
        data.append((row[0], row[1]))

```
